```python
sales = float(input("Enter sales ($): "))
print(f"Sales: ${sales}")
COGS = sales * 0.4
print(f"Cost of Goods Sold: ${COGS}")
GP = sales - COGS
print(f"Gross Profit: ${GP}")

TaxRate = float(input("Enter Tax Rate (%): "))
taxes = (GP * TaxRate/100)
print(f"Taxes: ${taxes}")
NI = GP - taxes
print(f"Net Income: ${NI}")
```

```python
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
op = input("Choose numerical operation (+, -, *, etc): ")

if op == "+":
  res = num1 + num2
elif op == "-":
  res = num1 - num2
elif op == "*":
  res = num1 * num2
elif op == "/":
  res = num1 / num2
elif op == "//":
  res = num1 // num2
elif op == "**":
  res = num1 ** num2
elif op == "%":
  res = num1 % num2
else:
  print("Invalid operation")

print(f"Result: {res}")
```

```python
print ("Hour",  "Fee")
for hour in range(1, 9, 1):
  fee = 5 + 2.5 * hour
  if fee < 10:
    fee = 10
  elif fee > 20:
    fee = 20
  print(f"{hour}     {fee}")
```
