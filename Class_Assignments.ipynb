```python
sales = float(input("Enter sales ($): "))
print(f"Sales: ${sales}")
COGS = sales * 0.4
print(f"Cost of Goods Sold: ${COGS}")
GP = sales - COGS
print(f"Gross Profit: ${GP}")

TaxRate = float(input("Enter Tax Rate (%): "))
taxes = (GP * TaxRate/100)
print(f"Taxes: ${taxes}")
NI = GP - taxes
print(f"Net Income: ${NI}")
```

```python
import random
check = input("Do you want to play? (y/n)").lower()

while check == 'y':
  player_move = input("Enter your move (rock, paper, scissors)").lower()
  computer_move = random.choice(['rock', 'paper', 'scissors'])

  print(f"Player: {player_move}")
  print(f"Computer: {computer_move}")

  if player_move == computer_move:
    print("Tie!")
  elif (
      player_move == 'rock' and computer_move == 'scissors' or
      player_move == 'paper' and computer_move == 'rock' or
      player_move == 'scissors' and computer_move == 'paper'
      ):
    print("You win!")

  else:
    print("Computer wins!")

  check = input("Do you want to play again? (y/n)")
  if check == 'n':
    print("Thank you for playing")
```

```python
def getValues(n):
  assert isinstance(n, int), "input variable 'n' must be an integer"
  assert n > 0
  items = []

  for i in range (n):
    items.append(float(input("Add a value to the list: ")))

  return items[2:]

def calcAverage(alist):
  assert len(alist) > 2
  assert isinstance(alist, list)
  try:
    return sum(alist) / len(alist)
  except Exception as e:
    print(f"Error: {e}")

  except:
    raise ValueError("Length of list must be greater than 2")

def main():
  n = int(input("Enter the list of items: "))
  alist = getValues(n)
  avg = calcAverage(alist)
  print(avg)
main()
```

```python
class Greeting:
  def __init__(self, first_name, last_name):
    self.first_name = first_name
    self.last_name = last_name

  def display(self, formal = False):
    if formal:
      print(f"{self.last_name}, {self.first_name}")
    else:
      print(f"{self.first_name}")

  def __str__(self):
    return "Instance of class 'Greeting'"

name = Greeting("Jane", "Doe")
print(name)
name.display()
name.display(formal=True)


```

```python
class Employee:
  def __init__(self, name, wage, hours):
    self.name = name
    self.__hourly_wage = wage
    self.hours = hours

  def get_wage(self):
    return self.__hourly_wage

  def set_wage(self, newWage):
    assert newWage > 0, "Wage can not be 0 or negative"
    self.__hourly_wage = newWage

  def calculate_pay(self):
    return self.__hourly_wage * self.hours

  def __str__(self):
    return f"Employee: {self.name}\nHourly wages: {self.__hourly_wage}"

class Manager(Employee):
  def __init__(self, name, wage, hours, bonus):
    super().__init__(name, wage, hours)
    self.bonus = bonus

  def calculate_pay(self):
    return (1 + self.bonus) * self.get_wage() * self.hours

employee = Employee("Bob", 50, 10)
manager = Manager("Alice", 100, 15, 0.2)
print(employee)
print(manager)
print()
print(f"Employee {employee.name}'s weekly salary: ${employee.calculate_pay()}")
print(f"Manager {manager.name}'s weekly salary: ${manager.calculate_pay()}, that includes a {manager.bonus * 100}% bonus")
```
